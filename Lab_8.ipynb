```python
import simpy
```

```python
import time
from datetime import datetime
```

```python
def time_exp(env):
  start=time.perf_counter()
  print('The starting time of the process is %s'%datetime.now())
  yield env.timeout(10)
  print('The ending time of the process is %s'%datetime.now())
  end=time.perf_counter()
  print('Process executed in %d units of time'%(end-start))
```

```python
env=simpy.Environment()
env.process(time_exp(env))
env.run()
```

```python
import simpy.rt as rt   # rt refers to Real Time
```

```python
env1=simpy.RealtimeEnvironment(initial_time=0,factor=2,strict=False)
env1.process(time_exp(env1))
env1.run()    #execution time will be the product of timeout and factor
```

```python
def slow_process(env):
  time.sleep(2)
  yield env.timeout(4)
```

```python
env2=rt.RealtimeEnvironment(factor=1,strict=False)
env2.process(slow_process(env2))
try:
  env2.run()
  print('Execution was smooth.')
except RuntimeError:
  print('The process is slow as compared to the Real Wall Clock.')
```
