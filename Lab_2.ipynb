```python
for i in range(1,5):
  for j in range(1,i+1):
    print(i,end=" ")
  print()
```

```python
n=int(input("Enter the number:"))
fac=1

while(n>=1):
  fac=fac*n
  n=n-1
if(n<0):
  print("Factorial of negative number does not exist.")
else:
  print("The factorial of the entered number is",fac)
```

```python
m=int(input("Enter the number:"))
fact=1
for i in range(1,m+1):
  fact=fact*i

if(m<0):
  print("Factorial of negative number does not exist.")
else:
  print("The factorial of the entered number is",fact)
```

```python
def facto(n):
  if(n==0):
    return 1
  else:
    return n*facto(n-1)


num=int(input("Enter the number:"))
if(num<0):
  print("Factorial of negative number does not exist.")
else:
  print("The factorial of the entered number is",facto(num))
```

```python
def fibo(n):
  if n==1:
    return 0
  elif n==2:
    return 1
  else:
    return fibo(n-1)+fibo(n-2)

nu=int(input("Enter the number of terms needed:"))
print('The Fibonacci Series of %d terms is : '%nu)

for i in range(1,nu+1):
  print('{}'.format(fibo(i)),end='\t')
```
