AIM: Getting started with Simpy using Car Process

```python
def car(env):
  while True:
    print('Start parking at %d'%env.now)
    parking_duration=5
    yield env.timeout(parking_duration)
    print('Start driving at %d'%env.now)
    trip_duration=10
    yield env.timeout(trip_duration)
```

```python
import simpy
env=simpy.Environment()
env.process(car(env))
env.run(until=100)
```
