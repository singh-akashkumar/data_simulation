AIM: To Make a Program in SIMPY for Process Interaction Using Interrupt

```python
def driver_arrival(env,car):
  yield env.timeout(2)
  car.action.interrupt()

class car:
  def __init__(self,env):
    self.env=env
    self.action=env.process(self.drive())
  
  def drive(self):
    while True:
      print('The car is in parking and is charging at %d'%self.env.now)
      parking_duration=6
      try:
        yield self.env.process(self.charge(parking_duration))
      except simpy.Interrupt:
        print("Let's go home!")

      print('The car is driving at %d'%self.env.now)
      driving_duration=10
      yield self.env.timeout(driving_duration)

  def charge(self,duration):
    yield self.env.timeout(duration)
```

```python
import simpy
env=simpy.Environment()
b=car(env)
env.process(driver_arrival(env,b))
env.run(until=40)
```
