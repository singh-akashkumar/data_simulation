```python
import simpy
```

```python
def car(env,name,bcs,driving_time,charge_time):
  yield env.timeout(driving_time)
  print('%s arrived at bcs at time %d'%(name,env.now))
  with bcs.request() as req:
    yield req
    print('%s started charging itself at time %d'%(name,env.now))
    yield env.timeout(charge_time)
    print('%s leaving the charging station at %d'%(name,env.now))
```

```python
env=simpy.Environment()
bcs=simpy.Resource(env,capacity=2)
```

```python
for i in range(4):
  env.process(car(env,'car_%d'%i,bcs,i*2,5))
```

```python
env.run()
```
