```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import io
from google.colab import files
```

```python
uploaded=files.upload()
```

```python
df=pd.read_csv(io.BytesIO(uploaded['weather.csv']))
```

```python
df.head()
```

```python
fig,ax=plt.subplots()
ax.plot(df['Month'],df['MaxTemp'],color='blue',marker='*',label='Max_Temp')
ax.plot(df['Month'],df['MinTemp'],color='red',marker='*',label='Min_Temp')
ax.plot(df['Month'],df['Rainfall'],color='green',marker='v',label='Rainfall')
ax.set_xlabel('Month')
ax.set_ylabel('Temperature')
ax.set_title('Min and Max Temp vs Month')
ax.legend(loc='best')
fig.set_size_inches([10,5])
fig.savefig('weather1.png')
plt.plot()
```

```python
fig,ax=plt.subplots(3,1,sharex=True)
ax[0].plot(df['Month'],df['MaxTemp'],color='b',marker='*',label='Max_Temp')
ax[1].plot(df['Month'],df['MinTemp'],color='r',marker='*',label='Min_Temp')
ax[2].plot(df['Month'],df['Rainfall'],color='g',marker='v',label='Rainfall')
ax[0].legend(loc='upper center')
ax[1].legend(loc='upper center')
ax[2].legend(loc='upper center')
fig.set_size_inches([10,7])
plt.plot()
```

```python
fig,ax=plt.subplots(2,2)
ax[0][0].plot(df['Month'],df['MaxTemp'],color='b',marker='*',label='Max_Temp')
ax[0][1].plot(df['Month'],df['MinTemp'],color='r',marker='*',label='Min_Temp')
ax[1][0].plot(df['Month'],df['Rainfall'],color='g',marker='v',label='Rainfall')
ax[1][1].plot(df['Month'],df['Sunshine'],color='magenta',marker='v',label='Sunshine')
ax[0][0].legend(loc='upper center')
ax[0][1].legend(loc='upper center')
ax[1][0].legend(loc='upper center')
ax[1][1].legend(loc='upper center')
fig.set_size_inches([20,7])
plt.style.use('classic')
plt.plot()
```

```python
fig,ax=plt.subplots()
ax.plot(df['Month'],df['MaxTemp'],color='blue',marker='*',label='Max_Temp')
ax.set_xlabel('Month')
ax.set_ylabel('Maximum Temperature')
ax.set_title('Min and Max Temp vs Month')
ax1=ax.twinx()
ax1.plot(df['Month'],df['MinTemp'],color='red',marker='*',label='Min_Temp')
ax1.set_ylabel('Minimum Temperature')
ax.legend()
ax1.legend(loc='upper center')
fig.set_size_inches([10,5])
plt.style.use('seaborn')
plt.plot()
```

```python
fig,ax=plt.subplots()
ax.bar(df['Month'],df['MaxTemp'],color='blue',label='Max_Temp')
ax.bar(df['Month'],df['MinTemp'],color='red',bottom=df['MaxTemp'],label='Min_Temp')
ax.bar(df['Month'],df['Rainfall'],color='green',bottom=df['MaxTemp']+df['MinTemp'],label='Rainfall')
ax.set_xlabel('Month')
ax.set_ylabel('Temperature')
ax.set_title('Min and Max Temp and Rainfall vs Month')
ax.legend(loc='best')
fig.set_size_inches([10,5])
plt.plot()
```
