```python
import matplotlib.pyplot as plt
```

```python
year=[1960,1970,1980,1990,2000,2010,2020]
pop=[3,3.4,4.3,5,5.6,6.3,7.1]
```

```python
plt.plot(year,pop,color='blue',marker='v',linestyle='--')
plt.title('Population vs Year plot')
plt.xlabel('Year')
plt.ylabel('Population')
plt.yticks([1,2,3,4,5,6,7])
plt.show()
```

```python
plt.scatter(year,pop,color='red',marker='*')
plt.title('Population vs Year plot')
plt.xlabel('Year')
plt.ylabel('Population')
plt.yticks([1,2,3,4,5,6,7])
plt.show()
```

```python
gdp=[4,4.5,5.2,5.6,6.1,6.5,7.7]

plt.plot(year,pop,color='blue',marker='v',label='Population')
plt.plot(year,gdp,color='red',marker='o',label='GDP')
plt.title('Population vs Year plot')
plt.xlabel('Year')
plt.ylabel('Population and GDP')
plt.yticks([1,2,3,4,5,6,7])
plt.legend(loc='lower right')
plt.show()
```
