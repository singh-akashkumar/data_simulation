```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('tips.csv')
df.head()
```

```python
sns.catplot(x='smoker',data=df,kind='count')
plt.show()
```

```python
sns.catplot(x='smoker',data=df,kind='count',hue='sex')
plt.show()
```

```python
my_palette=sns.color_palette('hls')
my_palette
```

```python
sns.catplot(x='smoker',data=df,kind='count',col='sex',palette=my_palette)
plt.show()
```

```python
sns.catplot(x='smoker',data=df,kind='count',row='sex',palette=sns.color_palette('Set2'))
plt.show()
```

```python
sns.catplot(x='smoker',data=df,kind='count',row='sex',palette=sns.color_palette('Set1'),order=['Yes','No'])
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=df,kind='bar')
plt.show()
# the vertical lines in the middle of the bars are of Confidence Interval
```

```python
sns.barplot(x='day',y='total_bill',data=df)
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=df,kind='bar',ci=None)
plt.show()
```

```python
sns.catplot(x='total_bill',y='day',data=df,kind='bar',ci=None)
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=df,kind='bar',ci=None,estimator=np.median)
plt.show()
```

```python
np.random.seed(0)
num2=np.random.randint(0,50,100)
print(num2)
```

Lab 17 - Plotting of Random Numbers

```python
df=pd.DataFrame(num2)
df.head()
```

```python
df.columns=['Number Generated']
```

```python
df
```

```python
df['Probability']=df['Number Generated']/len(df)
```

```python
df
```

```python
plt.figure(figsize=(20,5))
plt.bar(df.index,df["Probability"])
plt.show()
```

```python
plt.figure(figsize=(25,5))
sns.barplot(x=df.index,y=df['Probability'],data=df)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(df['Number Generated'],df["Probability"])
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=df['Number Generated'],y=df['Probability'],data=df)
plt.show()
```
