```python
df1=pd.read_excel('data_3rd.xlsx',sheet_name=0)
df1.head()
```

```python
sns.relplot(x='Age',y='Income',data=df1,kind='line')
plt.show()
```

```python
df1['Income'].value_counts()
```

```python
sns.relplot(x='Age',y='Work_hour',data=df1,kind='line')
plt.show()
#the shadow type portion is of Confidence Interval
```

```python
sns.relplot(x='Age',y='Work_hour',data=df1,kind='scatter',hue='Gender')
plt.show()
# Note: relplot contains attribute aplha to specify the transparency of the plotting
```

```python
sns.relplot(x='Age',y='Work_hour',data=df1,kind='line',hue='Gender')
plt.show()
```

```python
sns.relplot(x='Age',y='Work_hour',data=df1,kind='line',col='Gender')
plt.show()
```

```python
sns.relplot(x='Age',y='Work_hour',data=df1,kind='line',col='Gender',ci='sd')
# ci stands for Confidence Interval and sd stands for Standard Deviation
plt.show()
```

```python
sns.relplot(x='Age',y='Work_hour',data=df1,kind='line',col='Gender',ci=None)
plt.show()
```
